```python
df = pd.DataFrame()

for subject in infos:
    print(subject)
    info = infos[subject]
    for r in info['_sessions']:
        se = pd.Series(name=info['_name_temp'].format(r))
        for k in info:
            if k.startswith('_'):
                continue
            if k.endswith('_temp'):
                se[k[:-5]] = info[k].format(r)
            else:
                se[k] = info[k]
        df = df.append(se)
df[df.isna()] = '--'
df
```

```python
# Check and mkdir sessions
# Toggle new_start switch
# new_start = True means restore old processed_data and build new processed_data empty folders
# new_start = False means keep old processed_data and adding none-exist session folders
new_start = True

# Check raw_path exists
for e in df['raw_path']:
    print(e)
    assert(os.path.exists(e))

# Save old processed_data dir
if new_start:
    if os.path.exists('processed_data'):
        os.rename('processed_data', 'processed_data_{}'.format(time.ctime()))

# Make processed_data dir
if not os.path.exists('processed_data'):
    os.mkdir('processed_data')

# Add sessions into processed_data dir
for e in df['processed_path']:
    print(e)
    if not os.path.exists(e):
        os.mkdir(e)
```
